```python
def data_gen(x,a,b,sigma):
    # z = np.random.normal(0, sigma, 200) # defines our Gaussian random variable
    noise = 1/(sigma*np.sqrt(2*np.pi))*np.exp(-(x)**2/(2*sigma**2)) #equation for gaussian?
    return a*x+b+noise
```

```python
def dot_product(x1, x2):
    x1 = np.array(x1)
    x2 = np.array(x2)
    
    product = x1@x2
    return product
```
